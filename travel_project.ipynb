```python
#project of travel agency form.
from tkinter import *
import mysql.connector
root=Tk()
root.title("Travel agency form")
root.geometry("655x333")


def getvalues():
    name=namevalue.get()
    phone=phonevalue.get()
    gender=gendervalue.get()
    Emergency=Emergencyvalue.get()
    Age=Agevalue.get()
    paymentmode=paymentmodevalue.get()
    foodservice=foodservicevalue.get()
    
    mydb=mysql.connector.connect(host='localhost',user='root',password='Sam@12345',database='travel_form')
    mycursur=mydb.cursor()
    try:
        sql = "INSERT INTO travel_form (name,phone,gender,Emergency,Age,paymentmode,foodservice) VALUE (%s,%s,%s,%s,%s,%s,%s)"
        val = (name,phone,gender,Emergency,Age,paymentmode,foodservice)
        mycursur.execute(sql,val)
        mydb.commit()
    except: 
        
        mydb.rollback()
        mydb.close()

    

travel=Label(root,text="Welcome to Travel agency.",font=("comicsansms 20 bold"),pady=20).grid(row=0,column=3)

name=Label(root,text="passenger name",font=("comicsansms 10 bold"))
phone=Label(root,text="Phone number",font=("comicsansms 10 bold"))
gender=Label(root,text="gender",font=("comicsansms 10 bold"))
Emergency=Label(root,text="Emergency number",font=("comicsansms 10 bold"))
Age=Label(root,text="Age",font=("comicsansms 10 bold"))
paymentmode=Label(root,text="Payment mode",font=("comicsansms 10 bold"))

name.grid(row=1,column=2)
phone.grid(row=2,column=2)
gender.grid(row=3,column=2)
Emergency.grid(row=4,column=2)
Age.grid(row=5,column=2)
paymentmode.grid(row=6,column=2)

#variable for storing values.
namevalue=StringVar()
phonevalue=StringVar()
gendervalue=StringVar()
Emergencyvalue=StringVar()
Agevalue=StringVar()
paymentmodevalue=StringVar()
foodservicevalue=IntVar()
#entry of form
nameentry=Entry(root,textvariable=namevalue).grid(row=1,column=3)
nameentry=Entry(root,textvariable=phonevalue).grid(row=2,column=3)
nameentry=Entry(root,textvariable=gendervalue).grid(row=3,column=3)
nameentry=Entry(root,textvariable=Emergencyvalue).grid(row=4,column=3)
nameentry=Entry(root,textvariable=Agevalue).grid(row=5,column=3)
nameentry=Entry(root,textvariable=paymentmodevalue).grid(row=6,column=3)


#checkbox and packing
foodservices=Checkbutton(text=" Want to prebook your meal ",variable=foodservicevalue).grid(row=7,column=3)

#button 

Button(text="submit to travel agency.",command=getvalues).grid(row=8,column=3)



root.mainloop()
```
